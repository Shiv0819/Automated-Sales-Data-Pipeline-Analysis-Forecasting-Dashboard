```python
import pandas as pd 
import numpy as np 
```

```python
customers = pd.read_csv('customers.csv')
products = pd.read_csv('products.csv')
sales = pd.read_csv('sale.csv')
payments = pd.read_csv('payment.csv')
```

```python
customers.head()
```

```python
customers.info()
```

```python
#missing null values
customers['age'].isnull().sum()
```

```python
# filling null values in age 
customers['age'] =customers['age'].fillna(customers['age'].mean())
```

```python
# change the datatypes
customers['age'] = customers['age'].astype(int)
```

```python
customers.columns
```

```python
# if there is any space in the name of columns
customers.columns = customers.columns.str.replace(' ','')
```

```python
# change the null values
customers['signup_date'].isnull().sum()
```

```python
# Filling missing value from date 
#most_commin_date = customers['signup_date'].mode()[0]
customers['signup_date'] = customers['signup_date'].fillna(customers['signup_date'].mode()[0])
```

```python
customers.isnull().sum()
```

```python
customers.dtypes
```

```python
#change the data type of date object to date
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
```

```python
customers.dtypes
```

```python
customers[['gender','city']].isnull().sum()
```

```python
customers['gender'] = customers['gender'].fillna(customers['gender'].mode()[0])
```

```python
customers['city'] = customers['city'].fillna(customers['city'].mode()[0])
```

```python
customers[['gender','city']].isnull().sum()
```
