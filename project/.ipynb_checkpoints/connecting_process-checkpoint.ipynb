```python
#########################THIS IS THE MAIN CODE TO CONNECTED THE PYTHON TO MYSQL
#import pandas as pd
#from sqlalchemy import create_engine
#import os

# Create SQLAlchemy engine
engine = create_engine(
    "mysql+pymysql://root:shiva1911@localhost/project"
)

# Folder containing CSV files
folder_path = r"C:\Users\User\Desktop\project"

# IMPORTANT: correct order (FK safe)
files = [
    ("customer.csv", "customers"),
    ("products.csv", "products"),
    ("sale.csv", "sales"),
    ("payment.csv", "payments")
]

for csv_file, table_name in files:
    file_path = os.path.join(folder_path, csv_file)
    
    print(f"Reading {csv_file}...")
    df = pd.read_csv(file_path)

    # Convert date columns safely
    for col in df.columns:
        if "date" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce")

    print(f"Inserting into {table_name}...")
    
    df.to_sql(
        name=table_name,
        con=engine,
        if_exists="append",   # DO NOT replace
        index=False,
        chunksize=10          # prevents kernel crash
    )

    print(f"‚úÖ {table_name} inserted successfully\n")

print("üéâ ALL DATA INSERTED SUCCESSFULLY")
```

```python
import pandas as pd
import pymysql

# 1Ô∏è‚É£ Connect to MySQL
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="shiva1911",
    database="project"
)

# 2Ô∏è‚É£ Your full SQL query
query = """
SELECT 
    c.customer_name,
    c.age,
    c.gender,
    c.city,
    c.signup_date,
    p.product_name,
    p.category,
    p.price,
    s.sale_year,
    s.quantity,
    pay.payment_mode,
    pay.payment_status,
    (p.price * s.quantity) AS total_sale
FROM sales s
JOIN customers c ON s.customer_id = c.customer_id
JOIN products p ON s.product_id = p.product_id
LEFT JOIN payments pay ON s.sale_id = pay.sale_id
ORDER BY c.customer_name, s.sale_year
LIMIT 22;
"""

# 3Ô∏è‚É£ Load data into Pandas DataFrame
df = pd.read_sql(query, conn)

# 4Ô∏è‚É£ Preprocessing (optional)
# Encode gender
df['gender'] = df['gender'].map({'Male':0, 'Female':1})

# Fill missing payment info
df['payment_mode'] = df['payment_mode'].fillna('Unknown')
df['payment_status'] = df['payment_status'].fillna('Pending')

# Calculate total_sale again if needed (already in SQL)
df['total_sale'] = df['total_sale'].fillna(0)

# Optional: convert signup_date to datetime
df['signup_date'] = pd.to_datetime(df['signup_date'])

# 5Ô∏è‚É£ Quick check
print(df.head())
print(df.info())
```
